```python
import pandas as pd
import numpy as np
```

#1) Generación del dataset

```python
# URL base del repositorio en GitHub (raw)
url_base = "https://raw.githubusercontent.com/Fran251184/pasantia_TUIA_CIM_Francisco_J._Alomar/main/csv_por_medio/"

# Definición de las rutas de los archivos CSV en GitHub
clarin = url_base + "clarin.csv"
elciudadano = url_base + "elciudadano.csv"
infobae = url_base + "infobae.csv"
lacapital = url_base + "lacapital.csv"
lanacion = url_base + "lanacion.csv"
rosario3 = url_base + "rosario3.csv"

# Crear un diccionario con los archivos y los nombres de los DataFrames
archivos = {
    'df_clarin': clarin,
    'df_elciudadano': elciudadano,
    'df_infobae': infobae,
    'df_lacapital': lacapital,
    'df_lanacion': lanacion,
    'df_rosario3': rosario3
}

# Crear los DataFrames e imprimir un mensaje de confirmación
dataframes = {}

for nombre_df, url in archivos.items():
    try:
        dataframes[nombre_df] = pd.read_csv(url)
        print(f"{nombre_df} cargado con éxito!")
    except Exception as e:
        print(f"Error al cargar {nombre_df}: {e}")
```

```python
# Declaración explícita de cada DataFrame a partir del diccionario
df_clarin = dataframes['df_clarin']
df_elciudadano = dataframes['df_elciudadano']
df_infobae = dataframes['df_infobae']
df_lacapital = dataframes['df_lacapital']
df_lanacion = dataframes['df_lanacion']
df_rosario3 = dataframes['df_rosario3']
```

```python
# Concatenar los DataFrames y resetear el índice
df_concatenado = pd.concat([df_clarin, df_elciudadano, df_infobae, df_lacapital, df_lanacion, df_rosario3], ignore_index=True)

# Normalizar la columna comentarios_fb (reemplazar valores vacíos por NaN)
df_concatenado['comentarios_fb'] = df_concatenado['comentarios_fb'].apply(
    lambda x: np.nan if str(x).strip() in ["[]", "[\"\"]", "['']"] else x
)

# Mostrar el resultado
df_concatenado
```
